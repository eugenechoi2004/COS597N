Okay, the plan is to fine tune on very small dataset. Let's get everything we need.

```python
dp_ids = []
seqs = []
labels = []
plddts = []
for i, line in enumerate(lines):
    if '>' in line:
        dp_ids.append(line.split('|')[1])
        seqs.append(lines[i+1])
        labels.append(lines[i+3])
        plddts.append(lines[i+5].split('|'))
```

```python
len(dp_ids)
```

```python
len(seqs)
```

```python
len(labels)
```

```python
full = [[float(i) for i in label] for label in labels]
```

```python
full[0]
```

```python
for i in range(len(dp_ids)):
    a = len(seqs[i])
    b = len(full[i])
    c = len(plddts[i])
    assert(a == b)
    assert(a == c)
```

```python
len(plddts)
```

```python
subset_df = pd.DataFrame({
    'disprot_ID': dp_ids,
    'Sequence': seqs,
    'full': full,
    'pLLDT': plddts
})
```
